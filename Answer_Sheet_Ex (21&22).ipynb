```python
"""
I am going to do the task of designing an object-oriented library management system in Python.
This system manages various aspects of a library, including management of books, staff, members, and reservations.
My goal is to create an interactive system that demonstrates the key principles of object-oriented programming.
My system should contain the following classes with specified properties and methods:

1.Person (Abstract Base Class)
    •Attributes: name,age
    •Methods:
        •Method to initialize the name and age.
        •edit_profileto modify the name and age.
        •An abstract method describe_role.

2.Librarian (Inherits from Person)
    •Attributes: librarian_id,name,age
    •Methods:
        •Method for create a unique ID for each librarian.
        •describe_role to print the role description.

3.Member (Inherits from Person)
    •Attributes: member_id, name,age, books_borrowed
    •Methods:
        •Method for create a unique ID for each member.
        •borrow_book and return_book to manage book borrowings.
        •describe_roleto print the role description.

4.Book
    •Attributes: title,author,book_id,is_borrowed
    •Methods:
        •Method for createa unique ID for each book.
        •check_availability to see if the book can be borrowed.
5.Library
    •Attributes: name, books, members, librarians, reservations
    •Methods:
        •Method to initialize the library.
        •Methods to add books, register members, hire librarians, manage reservations.

6.Reservation
    •Attributes: reservation_id ,member ,book,  reservation_date ,status
    •Methods:
        •Method to create a new reservation.
        •cancel_reservationto cancel an existing reservation.

"""

# write code here:
#-----------------------------   class Person   ------------------------------
class Person:
  def __init__(self, name, age):
    self._name = name 
    self._age = age 

  # Getter method
  def get_name(self):
    return self._name
  
  # setter method
  def set_name(self, name):
    self._name = name

  # Getter method
  def get_age(self):
    return self._age

  # setter method
  def set_age(self, age):
    self._age = age

  # Method to edit the profile
  def edit_profile(self, name, age):
    self.set_name(name)
    self.set_age(age)

  # An abstract method to describe the role
  def describe_role(self):
    pass

#-----------------------------   class  Librarian   ----------------------------
class Librarian(Person):
  librarian_id=1000    
  def __init__(self, name, age, librarian_id):
    super().__init__(name, age) 
    self._librarian_id = librarian_id 

  # create a unique ID for each librarian
  @staticmethod
  def create_id():
        id1 = "L"+str(Librarian.librarian_id)
        Librarian.librarian_id += 1
        return id1

  # Method to describe the role
  def describe_role(self):
    print(f"I am a librarian. My name is {self.get_name()} and my ID is {self._librarian_id}.")


#-----------------------------   class  Member   ----------------------------
class Member(Person):
  member_id = 1000 
  def __init__(self, name, age, member_id, books_borrowed):
    super().__init__(name, age) 
    self._member_id = member_id 
    self._books_borrowed = books_borrowed 

  # create a unique ID for each member
  @staticmethod
  def create_id():
      id1 = "M"+str(Member.member_id)
      Member.member_id += 1
      return id1

  # borrow a book
  def borrow_book(self, book):
    self._books_borrowed.append(book)

  # return a book
  def return_book(self, book):
    self._books_borrowed.remove(book)

  # describe the role
  def describe_role(self):
    print(f"I am a member. My name is {self.get_name()} and my ID is {self._member_id}. I have borrowed {len(self._books_borrowed)} books.")

#-----------------------------   class Book   ----------------------------
class Book:
  book_id = 1000 
  def __init__(self, title, author, book_id, is_borrowed):
    self._title = title 
    self._author = author 
    self._book_id = book_id 
    self._is_borrowed = is_borrowed 
    

  # create a unique ID for each book
  @staticmethod
  def create_id():
      id1 = "B"+str(Book.book_id)
      Book.book_id += 1
      return id1

  # check the availability of the book
  def check_availability(self):
    return not self._is_borrowed

#-----------------------------   class Library   ----------------------------
class Library:
  
  def __init__(self, name, books, members, librarians, reservations):
    self._name = name 
    self._books = books 
    self._members = members 
    self._librarians = librarians 
    self._reservations = reservations 

  # initialize the library
  @staticmethod
  def initialize():
    librarian = Librarian("Ali", 35, Librarian.create_id())   # librarian 
    library = Library("My Libraray",[],[],librarian,None)
    return library

  # add a book to the library
  def add_book(self, book):
    self._books.append(book)

  # register a member to the library
  def register_member(self, member):
    self._members.append(member)

  # hire a librarian to the library
  def hire_librarian(self, librarian):
    self._librarians.append(librarian)

  # manage a reservation
  def manage_reservation(self, reservation):
    self._reservations.append(reservation)
    
  
# This method is an addition to the practice, and the test and printing of the books available in the library is written
  def print_datails(self):
    print(f"Library name: {self._name}")

    for book in self._books:
      print(f"Title: {book._title}")
      print(f"Author: {book._author}")
      print(f"ID: {book._book_id}")
      print(f"Available: {book.check_availability()}")
      print() # A blank line for readability

    # details of the members
    print("Members:")
    for member in self._members:
      print(f"Name: {member.get_name()}")
      print(f"Age: {member.get_age()}")
      print(f"ID: {member._member_id}")
      print(f"Books borrowed: {len(member._books_borrowed)}")
      print() 


#-----------------------------   class Reservation   ----------------------------
class Reservation:
  reservation_id  = 1000 
  def __init__(self, reservation_id, member, book, reservation_date, status):
    self._reservation_id = reservation_id 
    self._member = member 
    self._book = book 
    self._reservation_date = reservation_date
    self._status = status

  # create a new reservation
  @staticmethod
  def create_reservation(member, book, reservation_date):
    # Some logic to generate a unique ID and a status
    id1 = "R"+str(Reservation.reservation_id)
    Reservation.reservation_id += 1
    return Reservation(id1, member, book, reservation_date, "Pending")

  # cancel a reservation
  def cancel_reservation(self):
    self._status = "Cancelled"

    
```

```python


# Creating a library
my_library = Library.initialize()


# Creating five of the most famous Python books in the field of data science
# Some famous books have been used
book1 = Book("Python for Data Analysis", "Wes McKinney", Book.create_id(), False)
book2 = Book("Python Data Science Handbook", "Jake VanderPlas", Book.create_id(), False)
book3 = Book("Hands-on Machine Learning with Scikit-Learn", "Aurelien Geron", Book.create_id(), False)
book4 = Book("Data Science from Scratch", "Joel Grus", Book.create_id(), False)
book5 = Book("Automate the Boring Stuff with Python", "Al Sweigart", Book.create_id(), False)

# Adding the books to the library
my_library.add_book(book1)
my_library.add_book(book2)
my_library.add_book(book3)
my_library.add_book(book4)
my_library.add_book(book5)

# Creating three members with desired names
member1 = Member("Alice", 20, Member.create_id(), [])
member2 = Member("Bob", 25, Member.create_id(), [])
member3 = Member("Charlie", 30, Member.create_id(), [])

my_library.print_datails()

# Registering the members to the library
my_library.register_member(member1)
my_library.register_member(member2)
my_library.register_member(member3)

# Using the defined functions for these members and these books and printing the necessary output
# Member 1 borrows book 1 and book 2
member1.borrow_book(book1)
member1.borrow_book(book2)

# Member 2 borrows book 3 and book 4
member2.borrow_book(book3)
member2.borrow_book(book4)

# Member 3 tries to borrow book 5 but it is not available
member3.borrow_book(book5)

# Member 1 returns book 1
member1.return_book(book1)

# Member 3 borrows book 1
member3.borrow_book(book1)

# Member 2 creates a reservation for book 5
reservation = Reservation.create_reservation(member2, book5, "2023-12-30")

# Member 1 cancels the reservation
reservation.cancel_reservation()

# Printing the role description of each member
member1.describe_role()
member2.describe_role()
member3.describe_role()
```
